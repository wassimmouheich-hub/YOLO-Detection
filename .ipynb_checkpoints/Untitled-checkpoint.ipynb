```python
import os
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.python.framework.ops import EagerTensor
from tensorflow.keras.models import load_model
%matplotlib inline
```

```python
def yolo_filter_boxes(boxes, box_confidence, box_class_probs, threshold = .6):
    x = 10
    y = tf.constant(100)
    box_scores = box_confidence * box_class_probs #P_c*C 
    box_classes = tf.math.argmax(box_scores, axis=-1)
    #permet de prendre le score de la classe la plus eleve,on l'envoie ensuite à tf.math.reduce_max(box_scores, axis=-1) pour garder que le score le plus élévé. 
    box_class_scores = tf.math.reduce_max(box_scores, axis=-1)#axis=-1 car c'est là que se trouve les scores des classes
    filtering_mask = box_class_scores >= threshold#permet de garder les boîtes qui sont que >= à threshold
    #tf.boolean_mask permet d'appliquer le filtre 
    scores = tf.boolean_mask(box_class_scores, filtering_mask)
    boxes = tf.boolean_mask(boxes, filtering_mask)
    classes = tf.boolean_mask(box_classes, filtering_mask)
    return scores, boxes, classes
```

```python
tf.random.set_seed(10)
box_confidence = tf.random.normal([19, 19, 5, 1], mean=1, stddev=4, seed = 1)
boxes = tf.random.normal([19, 19, 5, 4], mean=1, stddev=4, seed = 1)
box_class_probs = tf.random.normal([19, 19, 5, 80], mean=1, stddev=4, seed = 1)
scores, boxes, classes = yolo_filter_boxes(boxes, box_confidence, box_class_probs, threshold = 0.5)
print("scores[2] = " + str(scores[2].numpy()))
print("boxes[2] = " + str(boxes[2].numpy()))
print("classes[2] = " + str(classes[2].numpy()))
print("scores.shape = " + str(scores.shape))
print("boxes.shape = " + str(boxes.shape))
print("classes.shape = " + str(classes.shape))
```

```python
def iou(box1, box2):
    (box1_x1, box1_y1, box1_x2, box1_y2) = box1
    (box2_x1, box2_y1, box2_x2, box2_y2) = box2
    xi1 = max(box1_x1, box2_x1)#coordonnée x la plus à droite
    yi1 = max(box1_y1, box2_y1)#coordonnée y la plus en bas
    xi2 = min(box1_x2, box2_x2)#coordonnée x la plus à gauche
    yi2 = min(box1_y2, box2_y2)#coordonnée y la plus haute
    #ces coordonnées nous permettent de trouver les coins du rectangle d'intersection
    #on a plus qu'à les soustraire 
    inter_width = xi2-xi1
    inter_height = yi2-yi1
    #si les boîtes ne se touchent pas,inter_width sera négatif,on utilise max(..,0) pour éviter la multiplication de 2 chiffres négatif(donne un chiffre positif) ce qui peut fausser le résultat
    inter_area = max(inter_width, 0) * max(inter_height, 0)#calcule de l'aire d'un rectangle
    box1_area = (box1_x2 - box1_x1) * (box1_y2 - box1_y1)
    box2_area = (box2_x2 - box2_x1) * (box2_y2 - box2_y1)
    union_area = box1_area + box2_area - inter_area#probabilité de A union B
    #iou est la division de A inter B sur  A union B
    iou = inter_area/union_area
    return iou
```

```python
## Test case 1: boxes intersect
box1 = (2, 1, 4, 3)
box2 = (1, 2, 3, 4)

print("iou for intersecting boxes = " + str(iou(box1, box2)))

## Test case 2: boxes do not intersect
box1 = (1,2,3,4)
box2 = (5,6,7,8)
print("iou for non-intersecting boxes = " + str(iou(box1,box2)))

## Test case 3: boxes intersect at vertices only
box1 = (1,1,2,2)
box2 = (2,2,3,3)
print("iou for boxes that only touch at vertices = " + str(iou(box1,box2)))

## Test case 4: boxes intersect at edge only
box1 = (1,1,3,3)
box2 = (2,3,3,4)
print("iou for boxes that only touch at edges = " + str(iou(box1,box2)))
```

```python
def yolo_non_max_suppression(scores, boxes, classes, max_boxes = 10, iou_threshold = 0.5):
    max_boxes_tensor = tf.Variable(max_boxes, dtype='int32')#maximum d'objet qu'on autorise à détecter
    nms_indices = tf.image.non_max_suppression(boxes, scores, max_boxes, iou_threshold, name=None)
    #la fonction tf.gather permet de rassembler ou récupérer les numéros de nms 
    scores = tf.gather(scores, nms_indices)
    boxes = tf.gather(boxes, nms_indices)
    classes = tf.gather(classes, nms_indices)
    #ps la fonction avant ne sert à rien
    return scores, boxes, classes
```

```python
tf.random.set_seed(10)
scores = tf.random.normal([54,], mean=1, stddev=4, seed = 1)
boxes = tf.random.normal([54, 4], mean=1, stddev=4, seed = 1)
classes = tf.random.normal([54,], mean=1, stddev=4, seed = 1)
scores, boxes, classes = yolo_non_max_suppression(scores, boxes, classes)

print("scores[2] = " + str(scores[2].numpy()))
print("boxes[2] = " + str(boxes[2].numpy()))
print("classes[2] = " + str(classes[2].numpy()))
print("scores.shape = " + str(scores.numpy().shape))
print("boxes.shape = " + str(boxes.numpy().shape))
print("classes.shape = " + str(classes.numpy().shape))
```

```python
#toutes les fonctions d'avant servent à éviter d'avoir trop de boîtes sur l'image
#cette fonction,en bas,permet d'avoir les coins de la boîtes
def yolo_boxes_to_corners(box_xy, box_wh):
    # Calcul des coins min (haut-gauche)
    box_mins = box_xy - (box_wh / 2.)
    
    # Calcul des coins max (bas-droite)
    box_maxes = box_xy + (box_wh / 2.)

    # On concatène les deux pour avoir (x1, y1, x2, y2) sur la dernière dimension (axis=-1)
    # box_mins[..., 0] est x1, box_mins[..., 1] est y1
    # box_maxes[..., 0] est x2, box_maxes[..., 1] est y2
    
    boxes = tf.concat([
        box_mins[..., 1:2],  # y1
        box_mins[..., 0:1],  # x1
        box_maxes[..., 1:2], # y2
        box_maxes[..., 0:1]  # x2
    ], axis=-1)

    return boxes
```

```python
#la fonction scale_boxe permet une remise à l'echelle(pour que les boîtes s'affichent au bon endroit,grâce à un produit en croix)
def scale_boxes(boxes, image_shape):
    # On récupère la hauteur et la largeur de l'image d'origine
    height = image_shape[0]
    width = image_shape[1]
    # On crée un vecteur de redimensionnement (K)
    dims = tf.stack([height, width, height, width])#permet de créer un vecteur
    #tf.stack permet d'empiler dans l'ordre donnée height etc
    #les coordonnées des boîtes sont en virgules,or tu veux les multiplier avec des nombres entiers(taille de ton image)
    #tf.cast permet de transformer des nombres entiers en nombres réelles pour permettre la multiplication
    dims = tf.cast(dims, dtype=tf.float32)
    #on change ensuite la forme du tableaux mais pas les données(utile pour le broadcasting)
    dims = tf.reshape(dims, [1, 4])
    # On multiplie les coordonnées normalisées (0 à 1) par les dimensions réelles
    boxes = boxes * dims
    return boxes
```

```python
def yolo_eval(yolo_outputs, image_shape = (720, 1280), max_boxes=10, score_threshold=.7, iou_threshold=.5):
    
    # On vérifie combien d'éléments on a reçu
    if len(yolo_outputs) == 4:
        # Cas du test manuel (4 variables séparées)
        box_xy, box_wh, box_confidence, box_class_probs = yolo_outputs
    elif len(yolo_outputs) == 3:
        # Cas du modèle réel (xy et wh sont collés dans le premier tenseur)
        box_outputs, box_confidence, box_class_probs = yolo_outputs
        box_xy = box_outputs[..., :2]
        box_wh = box_outputs[..., 2:]
    else:
        raise ValueError("YOLO outputs doit contenir 3 ou 4 éléments.")

    boxes = yolo_boxes_to_corners(box_xy, box_wh)#on récupére les coordonnées des coins
    scores, boxes, classes = yolo_filter_boxes(boxes, box_confidence, box_class_probs, score_threshold)
    boxes = scale_boxes(boxes, image_shape)
    scores, boxes, classes = yolo_non_max_suppression(scores, boxes, classes, max_boxes, iou_threshold)
    return scores, boxes, classes
```

```python
tf.random.set_seed(10)
yolo_outputs = (tf.random.normal([19, 19, 5, 2], mean=1, stddev=4, seed = 1),
                tf.random.normal([19, 19, 5, 2], mean=1, stddev=4, seed = 1),
                tf.random.normal([19, 19, 5, 1], mean=1, stddev=4, seed = 1),
                tf.random.normal([19, 19, 5, 80], mean=1, stddev=4, seed = 1))
scores, boxes, classes = yolo_eval(yolo_outputs)
print("scores[2] = " + str(scores[2].numpy()))
print("boxes[2] = " + str(boxes[2].numpy()))
print("classes[2] = " + str(classes[2].numpy()))
print("scores.shape = " + str(scores.numpy().shape))
print("boxes.shape = " + str(boxes.numpy().shape))
print("classes.shape = " + str(classes.numpy().shape))
```

```python
def read_classes(classes_path):
    with open(classes_path, "r") as f:
        # On lit toutes les lignes, on enlève les espaces/retours à la ligne
        class_names = [line.strip() for line in f.readlines()]
    return class_names
```

```python
def read_anchors(anchors_path):
    with open(anchors_path, "r") as f:
        # On récupère la ligne de texte
        anchors_txt = f.readline()
        # On sépare par les virgules et on convertit en nombres décimaux
        anchors = [float(x) for x in anchors_txt.split(',')]
    
    # On les regroupe par paires (largeur, hauteur)
    # .reshape(-1, 2) crée un tableau de N lignes et 2 colonnes
    return np.array(anchors).reshape(-1, 2)
```

```python
#on va implémenter la fonction de prétraitement d'image
#le modele ne peut pas lire les fichiers JPEG,il a donc besoin d'un tenseur de chiffres,de redimensionné à la taille attendue et normalisé
from PIL import Image
def preprocess_image(img_path, model_image_size):
    # 1. Ouvrir l'image
    image = Image.open(img_path)

    # 2. Redimensionner l'image pour le modèle (ex: 608x608)
    resized_image = image.resize(tuple(reversed(model_image_size)), Image.BICUBIC)

    # 3. Convertir en tableau numpy et normaliser (0 à 1)
    image_data = np.array(resized_image, dtype='float32')
    image_data /= 255.

    # 4. Ajouter une dimension pour le "batch" (ex: de (608,608,3) à (1,608,608,3))
    image_data = np.expand_dims(image_data, 0)

    return image, image_data
```

```python
from PIL import ImageDraw, ImageFont
import colorsys
def draw_boxes(image, out_scores, out_boxes, out_classes, class_names, colors):
    #Dessine les boîtes de détection sur l'image originale.
    draw = ImageDraw.Draw(image)
    # Choix d'une police
    try:
        font = ImageFont.truetype(font='font/FiraMono-Medium.otf', size=np.floor(3e-2 * image.size[1] + 0.5).astype('int32'))
    except:
        font = ImageFont.load_default()

    thickness = (image.size[0] + image.size[1]) // 300

    for i, c in enumerate(out_classes):
        predicted_class = class_names[c]
        box = out_boxes[i]
        score = out_scores[i]

        label = '{} {:.2f}'.format(predicted_class, score)
        
        # Coordonnées de la boîte
        top, left, bottom, right = box
        top = max(0, np.floor(top + 0.5).astype('int32'))
        left = max(0, np.floor(left + 0.5).astype('int32'))
        bottom = min(image.size[1], np.floor(bottom + 0.5).astype('int32'))
        right = min(image.size[0], np.floor(right + 0.5).astype('int32'))

        # Dessiner le rectangle
        for j in range(thickness):
            draw.rectangle([left + j, top + j, right - j, bottom - j], outline=colors[c])

        # Dessiner l'étiquette (label)
        label_size = draw.textbbox((0, 0), label, font=font)
        label_origin = np.array([left, top - (label_size[3] - label_size[1])])
        
        draw.rectangle([tuple(label_origin), left + (label_size[2] - label_size[0]), top], fill=colors[c])
        draw.text(tuple(label_origin), label, fill=(0, 0, 0), font=font)
    return image
```

```python
# Générer des couleurs pour les classes (80 classes pour COCO)
hsv_tuples = [(x / len(class_names), 1., 1.) for x in range(len(class_names))]
colors = list(map(lambda x: colorsys.hsv_to_rgb(*x), hsv_tuples))
colors = list(map(lambda x: (int(x[0] * 255), int(x[1] * 255), int(x[2] * 255)), colors))
```

```python
def yolo_head(feats, anchors, num_classes):
    # 1. On s'assure d'avoir le tenseur (pas la liste)
    if isinstance(feats, (list, tuple)):
        feats = feats[0]
    
    # 2. On récupère le nombre d'anchors localement
    n_anchors = len(anchors) # Doit être = à 3 pour le modèle actuel
    
    # 3. On récupère la shape dynamique
    # feats shape est (1, 19, 19, 255)
    shape = tf.shape(feats)
    grid_h, grid_w = shape[1], shape[2]
    
    # 4. LE RESHAPE CORRIGÉ
    # Au lieu d'utiliser num_classes qui vient d'ailleurs, 
    # on laisse TF calculer la dernière dimension (85)
    # 255 / 3 = 85.
    feats = tf.reshape(feats, [-1, grid_h, grid_w, n_anchors, 85])

    # 5. On sépare les résultats
    box_xy = tf.sigmoid(feats[..., :2])
    box_wh = tf.exp(feats[..., 2:4])
    box_confidence = tf.sigmoid(feats[..., 4:5])
    box_class_probs = tf.nn.softmax(feats[..., 5:])

    # 6. Création de la grille pour repositionner les boîtes
    grid_y = tf.tile(tf.reshape(tf.range(grid_h), [-1, 1, 1, 1]), [1, grid_w, 1, 1])
    grid_x = tf.tile(tf.reshape(tf.range(grid_w), [1, -1, 1, 1]), [grid_h, 1, 1, 1])
    grid = tf.cast(tf.concat([grid_x, grid_y], axis=-1), tf.float32)

    # 7. Finalisation des coordonnées
    box_xy = (box_xy + grid) / tf.cast(shape[1:3][::-1], tf.float32)
    
    anchors_tensor = tf.reshape(tf.constant(anchors, dtype=tf.float32), [1, 1, 1, n_anchors, 2])
    box_wh = (box_wh * anchors_tensor) / tf.cast(shape[1:3][::-1], tf.float32)

    return box_xy, box_wh, box_confidence, box_class_probs
```

```python
def predict(image_file):
    # 1. Prétraitement de l'image
    image, image_data = preprocess_image(image_file, model_image_size = (608, 608))

    # 2. Passer l'image dans le modèle (Forward Propagation)
    # yolo_model est le modèle chargé avec load_model
    yolo_model_outputs = yolo_model.predict(image_data)

    yolo_outputs = yolo_head(yolo_model_outputs, anchors_v3, len(class_names))
    
    # 3. Transformer les sorties brutes en boîtes filtrées (notre fonction yolo_eval !)
    # On passe les sorties du modèle à yolo_eval
    out_scores, out_boxes, out_classes = yolo_eval(yolo_outputs, [image.size[1], image.size[0]])

    # 4. Affichage des résultats
    print(f'Found {len(out_boxes)} boxes for {image_file}')

    # 5. Dessiner les boîtes sur l'image originale
    # On utilise la fonction draw_boxes qu'on a faite juste avant
    draw_boxes(image, out_scores, out_boxes, out_classes, class_names, colors)

    # 6. Sauvegarder et afficher l'image finale
    image.save(os.path.join(image_file), quality=90)

    # Pour afficher l'image directement dans le notebook :
    plt.imshow(image)
    plt.show()

    return out_scores, out_boxes, out_classes
```
